```python
import torch
import clip
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import tensorflow as tf
import pandas as pd

device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)

#folder1内のすべての画像ファイルを読み込み、PILで開く
import os

text_list = ["red","blue","yellow","white","home","station","house","building","tower","human","railroad","road","car","train","sky"]
text = clip.tokenize((f"a photo of a {c}") for c in text_list).to(device)

sim = []
folder_path = "/content/drive/MyDrive/Colab Notebooks/img2306"
for filename in os.listdir(folder_path):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        img_path = os.path.join(folder_path, filename)
        with Image.open(img_path).convert("RGB") as img:
            # Do something with the image
            #img_file = R"/content/0.jpg"
# リサイズ前の画像を読み込み
#img = Image.open(img_file).convert("RGB")
            #plt.imshow(np.array(img))

            image = preprocess(img).unsqueeze(0).to(device)

            with torch.no_grad():
                #plt.imshow(img)
                #img
                image_features = model.encode_image(image)
                text_features = model.encode_text(text)

    # 推論
    #logits_per_image, logits_per_text = model(image, text)
    #probs = logits_per_image.softmax(dim=-1).cpu().numpy()

# Pick the top 10 most similar labels for the image
                image_features /= image_features.norm(dim=-1, keepdim=True)
                text_features /= text_features.norm(dim=-1, keepdim=True)
                print(image_features.shape, text_features.shape)
                similarity = (100.0 * image_features @ text_features.T).softmax(dim=-1)

                sim.append(similarity.numpy().flatten())

                #print(similarity)
                #print(f"numpy = {similarity.numpy()}")

                values, indices = similarity[0].topk(10)
                #values, indices = similarity[0]
#print(image_features)
# 類似率の出力
#print("Label probs:", probs)

                print(f"\nTop predictions:{filename}\n")
                for value, index in zip(values, indices):
                    print(f"{text_list[index]:>16s}: {100 * value.item():.2f}%")
print(sim)

df = pd.DataFrame(sim)
#df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/img2306/tensor.csv', header=None)
writer = pd.ExcelWriter('/content/drive/MyDrive/Colab Notebooks/img2306/photo_features.xlsx')
df.to_excel(writer, sheet_name='Sheet1')
writer.save()
```

```python
%matplotlib nbagg
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from somoclu import Somoclu
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
#Colab上でアニメーション表示(javascript)
from matplotlib import rc
from IPython.display import HTML
import random
```

```python
#excelデータを読み込む
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/img2306/photo_features.xlsx", sheet_name=0, index_col=0)
X = df.values
#データを0-1に正規化
#scaler = MinMaxScaler()
#X_s = scaler.fit_transform(X)
#print(df.values)
row = X.shape[0]
col = X.shape[1]
print(row, col, X)
df_ind = df.index
print(df_ind)
```

```python
# 初期設定
T = 250  # 学習の総回数 T
sigma_max = 5.0  # バンド幅の最大値
sigma_min = 0.01  # バンド幅の最小値
learning_rate_max = 0.8  # 学習率の最大値
learning_rate_min = 0.05  # 学習率の最小値

n_row = 50 # 自己組織化マップの行数
n_col= 50 # 自己組織化マップの列数

# 近傍関数
def neighborhood_func(distance, sigma):
    return np.exp(-(distance**2) / (2*(sigma**2)))

# 学習率
def learning_rate(t):
    return learning_rate_max * ((learning_rate_min / learning_rate_max) ** (t / T))

# バンド幅
def sigma(t):
    return sigma_max * ((sigma_min / sigma_max) ** (t / T))

# Step5 繰り返し処理
# アニメーション用変数定義
fig = plt.figure()

colors = np.random.rand(n_row, n_col, 3)  # 初期のカラーマップRGB値
imgs = []
img = plt.imshow(colors)
imgs.append([img])

plotwin = [[0 for k in range(3)] for i in range(n_row)]

#import pdb

def somnew(n_row, n_col, feature3d, T):
    t = 0
    ind = 0
    Ir = learning_rate_max
    while (Ir > learning_rate_min) and (t < T):

        # Step2 全てのセルとの距離を計算
        distances = np.full((n_row, n_col, 3), 100.0)

        #for i in range(n_row):
            #for j in range(n_col):
        for i in random.sample(range(n_row), n_row):
            for j in random.sample(range(n_col), n_col):
                distances[i][j] = np.linalg.norm(feature3d - colors[i][j])

        # Step2 最小距離のセルを取得
        #pdb.set_trace()
        winner = np.unravel_index(np.argmin(distances), distances.shape)
        print(f"winner = {winner}")
        #breakpoint()

        # Step3 近傍関数を使いRGB値を更新(各セルの重みを更新)
        #PCAを用いた
        #for i in range(n_row):
            #for j in range(n_col):
        for i in random.sample(range(n_row), n_row):
            for j in random.sample(range(n_col), n_col):

                dist_to_winner = np.linalg.norm(np.array([i, j, 0]) - np.array(winner))
                h = neighborhood_func(dist_to_winner, sigma(t))
                lr = learning_rate(t)

                colors[i][j] += h * lr * (feature3d - colors[i][j])
                #print(f'learning rate {t} = {lr}') 学習率の収束状況の確認
        t += 1
    return winner

pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)

# 画像50枚分処理
for gazo in range(50):
    winner = somnew(50, 50, X_pca[gazo], T)
    plotwin[gazo][0] = winner[0]
    plotwin[gazo][1] = winner[1]
    plotwin[gazo][2] = str(gazo)
    img = plt.imshow(colors)
    # 連続描画するため画像データを蓄積
    imgs.append([img])

# Step4 カラーマップを可視化
for gazo in range(50):
    plt.plot(plotwin[gazo][0], plotwin[gazo][1], "o")
    plt.text(plotwin[gazo][0], plotwin[gazo][1], plotwin[gazo][2], va = 'top' , size = "small" )

ani = animation.ArtistAnimation(fig, imgs, interval=100)
ani.save('Self_org_m50.gif', writer='pillow')

# Colab上でアニメーション表示
rc('animation', html='jshtml')
ani
```

```python
#ライブラリを使ってみた
# SOMの定義
n_rows = 50
n_cols = 50
som = Somoclu(n_rows=n_rows, n_columns=n_cols,
                  initialization="pca", verbose=2)

# 学習
som.train(data=X, epochs=1000)

# U-matrixをファイル出力
colors = ["red"] * 50
colors.extend(["green"] * 50)
colors.extend(["blue"] * 50)
som.view_umatrix(labels=df_ind, bestmatches=True,bestmatchcolors=colors,
                     filename="umatrix.png")
```
